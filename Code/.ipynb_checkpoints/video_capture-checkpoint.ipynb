```python
import cv2 
import numpy as np 
import dlib 
import os

filename = 'vid.mp4'
video_images = 'video-images\\'
id=0
total_frame=0
frameFrequency=24
res = '720p'

def change_res(cap,w,h):
    cap.set(3,w)
    cap.set(4,h)

    
STD_DIMENSIONS =  {
    "480p": (640, 480),
    "720p": (1280, 720),
    "1080p": (1920, 1080),
    "4k": (3840, 2160),
}

def get_dims(cap, res='1080p'):
    width, height = STD_DIMENSIONS["480p"]
    if res in STD_DIMENSIONS:
        width,height = STD_DIMENSIONS[res]
    change_res(cap, width, height)
    return width, height

VIDEO_TYPE = {
    'avi': cv2.VideoWriter_fourcc(*'XVID'),
    'mp4': cv2.VideoWriter_fourcc(*'XVID'),
}

def get_video_type(filename):
    filename, ext = os.path.splitext(filename)
    if ext in VIDEO_TYPE:
        return  VIDEO_TYPE[ext]
    return VIDEO_TYPE['avi']


cap = cv2.VideoCapture(0)  
detector = dlib.get_frontal_face_detector()
vtype=get_video_type(filename)
dims=get_dims(cap, res)
out = cv2.VideoWriter(filename,vtype,6,dims)
i=0
while True: 
    prev_i=i
    ret, frame = cap.read() 
    out.write(frame)
    frame = cv2.flip(frame, 1) 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    faces = detector(gray) 
    i = 0
    for face in faces: 
        x, y = face.left(), face.top() 
        x1, y1 = face.right(), face.bottom() 
        cv2.rectangle(frame, (x, y), (x1, y1), (0, 255, 0), 2) 
        i = i+1
        cv2.putText(frame, 'face num'+str(i), (x-10, y-10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2) 
    
    if(i>1 or i<1):
        #print(i)
        prev_i=i
        total_frame += 1
        #if total_frame%frameFrequency == 0:
        id+=1
        image_name = video_images + str(id) +'.jpg'
        cv2.imwrite(image_name, frame)
        print(image_name)
    cv2.imshow('frame', frame) 
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release() 
cv2.destroyAllWindows()
```
