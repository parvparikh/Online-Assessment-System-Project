```python
import pymysql
import sys
import random
#database connection
connection = pymysql.connect(host="localhost", user="root", passwd="", database="exam")
cursor = connection.cursor()

remember='select * from qbank where bt="remember" and cid="CC101" and tid=1'
cursor.execute(remember)
rem_que=cursor.fetchall()

analyse='select * from qbank where bt="analyse" and cid="CC101" and tid=1'
cursor.execute(analyse)
ana_que=cursor.fetchall()


apply='select * from qbank where bt="apply" and cid="CC101" and tid=1'
cursor.execute(apply)
app_que=cursor.fetchall()


marks_rem=0
for i in range(len(rem_que)):
    marks_rem=marks_rem + rem_que[i][3]

marks_ana=0
for i in range(len(ana_que)):
    marks_ana=marks_ana + ana_que[i][3]

marks_app=0
for i in range(len(app_que)):
    marks_app=marks_app + app_que[i][3]


print("-------------------------------------------------------------------------------------------------------------------------")
print("Total questions of each class")
print()
print()
print("You have added "+str(len(rem_que))+" questions of class Remember with total marks = "+str(marks_rem))
print()
print("You have added "+str(len(ana_que))+" questions of class Analyse with total marks = "+str(marks_ana))
print()
print("You have added "+str(len(app_que))+" questions of class Apply with total marks = "+str(marks_app))
print()
print()
num_rem=int(input("How many questions of class Remember= "))

if num_rem > len(rem_que):
    print()
    print("You dont have enough questions of class Remember")
    sys.exit()


print()
print()

num_ana=int(input("How many questions of class Analyse= "))

if num_ana > len(ana_que):
    print()
    print("You dont have enough questions of class Analyse")
    sys.exit()
print()
print()
num_app=int(input("How many questions of class Apply= "))

if num_app > len(app_que):
    print()
    print("You dont have enough questions of class Apply")
    sys.exit()
    
print()
print()    
num_test=int(input("How many tests, you want to generate?")) 
    
    
quiz_arr=[] 
cnt=1
for i in range(0,num_test):
    
    print()
    print("--------------------------------------------------------------------------------------------------------------------")
    print("Test for student "+str(cnt))
    
    rem_arr=[]
    app_arr=[]
    ana_arr=[]
    
    
    r_num = random.sample(range(0, len(rem_que)), num_rem)
    for i in r_num:
        rem_arr.append(rem_que[i][4])
    
    
    a_num = random.sample(range(0, len(app_que)), num_app)
    for i in a_num:
        app_arr.append(app_que[i][4])
    
    
    an_num = random.sample(range(0, len(ana_que)), num_ana)
    for i in an_num:
        ana_arr.append(ana_que[i][4])
    
    print()
    print("---------------------------Remember----------------------------")
    
    for i in rem_arr:
        print()
        print(i)
        print()
    print()
  
    print("---------------------------Apply----------------------------")
    for i in app_arr:
        print()
        print(i)
        print()
    print()
  
    print("---------------------------Analyse----------------------------")
    for i in ana_arr:
        print()
        print(i)
        print()
    print()
   
    print("--------------------------------------------------------------------------------------------------------------------")
    cnt=cnt+1
```
